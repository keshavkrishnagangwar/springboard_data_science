## Part 1 - EDA

```python
display(data.describe()), display(data.info());
```

The first thing to do is to resample the data into 15 minute intervals.

```python
# use login as index
logins = pd.Series(data=1, index=data.login_time)
# resample to 15 min intervals
logins = logins.resample('15T').sum()
```

```python
logins.head()
```

Next, let's look at the distribution of logins over the time range.

## Monthly logins

```python
ax = logins.resample('M').sum().plot(kind='line')
ax.set_ylabel('Count');
```

We can see that logins peaked in March. For April, logins are only recorded until the 13th of the month. Let's take a closer look at the weekly logins over the date range.

## Weekly logins

```python
ax = logins.resample('W').sum().plot(kind='line')
ax.set_ylabel('Count');
```

From here we can see that logins are increasing on average over the date range. There is a slight decline at the start of April but it appears to be within a normal range of variance.

## Daily logins

```python
ax = logins.resample('D').sum().plot(kind='line')
ax.set_ylabel('Count');
```

The frequency of logins over the week looks consistent across the different months, suggesting that day of the week interacts with login counts.

## Logins by day of the week 

```python
# pass series to dataframe
login_df = logins.to_frame(name='num')
# get day of week
login_df['day_of_week'] = login_df.index.day_name()
```

```python
login_df.head()
```

```python
# aggregate by day of week
weekday = pd.pivot_table(login_df, index='day_of_week', aggfunc=np.sum)
# sort index
weekday = weekday.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
# visualize distribution
weekday.plot(kind='bar');
```

Here we can see that logins increase over the week and reach a peak on Saturday.

## What happened in March?

Looking at the plot of daily activity over the date range, there appears to be a portion of March that has a different pattern than the rest of the distribution.

```python
ax = logins.resample('D').sum().plot(kind='line')
ax.set_ylabel('Count');
```

Now, let's take a closer look at March.

```python
# extract March for evaluation
march = login_df[login_df.index.month == 3]
```

```python
ax = march.resample('D').sum().plot(kind='line')
ax.set_ylabel('Count');
```

```python
# extract week for closer inspection
weird_week = march[(march.index.day >= 16) & (march.index.day < 23)]
```

```python
weird_week.resample('D').plot(kind='line');
```

There appears to be an anomoly during this week, where logins during the weekday are similar to the weekend. To test this, let's extract the data from the preceeding week and compare logins. The preceeding week appears to be typical of a weekly pattern, so this will give us a sense if there is in fact a difference.

```python
# extract preceeding week for comparison
not_weird_week = march[(march.index.day >= 9) & (march.index.day < 16)]
```

```python
from scipy import stats
stats.ttest_ind(weird_week.num.values, not_weird_week.num.values)
```

With a p-value < 0.0001, there does appear to be a deviation during that week.

### Summary 

Starting on March 17, there appears to be a deviation in the typical pattern of logins. This is characterized by weekday login patterns that are similar to the weekend. This suggests that there are either (1) data quality issues where logins were not properly recorded, or (2) that something occured, resulting in higher than normal logins. This was supported by a t-test showing statistically significant increases in weekly logins.
